```python
#LABEL and ORDINAL ENCODER in scikit learn

#if the given data set is having the categorical data then convert it into  the numerical data
"""
Nominal data-->In the data set the output variable is categorical like(T/F) then the (LABEL Encoder) is used to convert 0 or 1.

In the input features  the categories are in order( Low, Medium, High	) then (ORDINAL Encoder) is used.

if catogeries are not in order like (names),(fruits-->apple,banana,etc) are not in order then (ONE HOT ENCODER) is used.

"""
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/CAR DETAILS FROM CAR DEKHO.csv')
df.head()
```

```python
df.drop('name',axis=1,inplace=True)
df
```

```python


"""
df['transmission'].map({'manual':0,'Automatic':1})
df['seller'].map({'individual':0,'dealer':1,'trustmark dealer':2})

#same thing we can do in python also
df['transmission'].apply(lambda x: 1 if x=='manual'else 0 )

"""
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_text=train_test_split(df.drop('fuel',axis=1),df['fuel'],test_size=0.2,random_state=42)

x_train.shape
```

```python
x_test.shape
```

```python
#converting the data into values in table by LABEL encoder

from sklearn.preprocessing import LabelEncoder,OrdinalEncoder

le=LabelEncoder()
y_train_scaled=le.fit_transform(y_train)
y_test_scaled=le.fit_transform(y_text)#this will give 0,1,2,.... to distinct names in column

y_train_scaled  # gives output as numpy array
```

```python
y_test_scaled
```

```python

#OrdinalEncoder

de=OrdinalEncoder()
x_train_scaled=de.fit_transform(x_train)
x_test_scaled=de.fit_transform(x_test)

x_train_scaled
```
