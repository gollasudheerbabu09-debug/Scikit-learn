```python
#Feature scaling (normalization and standardization)
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/CAR DETAILS FROM CAR DEKHO.csv')
df.head()
```

```python
#Normalization--(MIN-MAX scalar)
#The purpose is to ensure that all features contribute equally to the model and to avoid the domination of features with larger values.
df_new=df[['selling_price','km_driven','year']]

from sklearn.model_selection import train_test_split

x=df_new.drop('selling_price',axis=1)
y=df_new['selling_price']

x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_text=train_test_split(x,y,test_size=0.33,random_state=42)
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar=MinMaxScaler()#three arguments(feature_range: Any = (0, 1), *, copy: bool = True, clip: bool = False)
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.fit_transform(x_test)

x_train_scaled
```

```python
x_test_scaled
```

```python
#make the obtained output from the normalization to dataframe
x_train_scaled=pd.DataFrame(data=x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(data=x_test_scaled,columns=x_test.columns)

x_train_scaled.head()
```
